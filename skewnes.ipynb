```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("titanic.csv")
```

```python
dataset.head(3)
```

```python
# grater then 0.123 mean posiyive skewnes 
dataset["Age"].skew()
```

```python
dataset["Age"].mean(),dataset["Age"].median(),dataset["Age"].mode()[0]
```

```python
sns.histplot(data=dataset, x="Age")
plt.show()
```

```python
# negative skewnes
data = np.random.normal(0, 100
, 100)
```

```python
df = pd.DataFrame({"x":data})
df["x"].skew()
```

```python
sns.histplot(x="x", data=df)
plt.show()
```
